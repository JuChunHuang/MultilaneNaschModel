```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
routes = df.groupby("Route_ID")
route5 = routes.get_group(5).reset_index()
route90 = routes.get_group(90).reset_index()
route405 = routes.get_group(405).reset_index()
route520 = routes.get_group(520).reset_index()
df
```

```python
sdv100 = np.array([909, 1743, 2708, 3627, 4479])
sdv50 = np.array([313, 626, 980, 1290, 1622])
sdv0 = np.array([222, 444, 666, 893, 1109])
```

```python
sdv100DECR_5 = []
sdv100INCR_5 = []
sdv50DECR_5 = []
sdv50INCR_5 = []
sdv0DECR_5 = []
sdv0INCR_5 = []
for i in range(route5.shape[0]):
    sdv100DECR_5.append(sdv100[route5["Number_of_Lanes_DECR_MP_direction"][i]-1])
    sdv100INCR_5.append(sdv100[route5["Number_of_Lanes_INCR_MP_direction"][i]-1])
    sdv50DECR_5.append(sdv50[route5["Number_of_Lanes_DECR_MP_direction"][i]-1])
    sdv50INCR_5.append(sdv50[route5["Number_of_Lanes_INCR_MP_direction"][i]-1])
    sdv0DECR_5.append(sdv0[route5["Number_of_Lanes_DECR_MP_direction"][i]-1])
    sdv0INCR_5.append(sdv0[route5["Number_of_Lanes_INCR_MP_direction"][i]-1])
```

```python
plt.figure(figsize=(15,6))
w = route5["startMilepost"]
y = route5["Average_daily_traffic_counts_Year_2015"]/24/2
plt.plot(w, y, drawstyle="steps-pre")
plt.plot(w, sdv100DECR_5, drawstyle="steps-pre")
plt.plot(w, sdv50DECR_5, drawstyle="steps-pre")
plt.plot(w, sdv0DECR_5, drawstyle="steps-pre")
plt.legend(["real situation", r'$P_{SDV}$ = 100%', r'$P_{SDV}$ = 50%', r'$P_{SDV}$ = 0%'])
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.3)
plt.title("Comparison of Traffic Flow with Interstate 5")
plt.xlabel("Milepost (DECR)")
plt.ylabel("Total Traffic Flow (vehicles/hour)")
```
